```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sarahbismuth/dataa/master/shhoting%20final%20project%20data%20.csv')
df
```

```python
#This dataset is composed of  48 columns which is a lot, eventhough most of them present interesting value,from now we will use only the column that are relevant to answer the question about the shooter profile
df1=df[['State','Shooter Sex','Average Shooter Age','Shooter Race','Type of Gun - General','Military Experience','School Related','History of Mental Illness - Detailed','History of Mental Illness - General','Relationship to Incident Location','Number of Victim Fatalities','City','Targeted Victim/s - General']]
df1
```

```python
#Change some column name to a more convenient name, here MI means mental Illness
df1.rename(columns={"Type of Gun - General": "Type of gun",'History of Mental Illness - Detailed':"MI detailed","History of Mental Illness - General":"General MI","Shooter Sex":"Shooter_sex","Average Shooter Age":"Shooter_age","Targeted Victim/s - General":"targeted victims"},inplace=True)
```

```python
#this will delete  NAn rows if they exists 
df1.dropna(how='all')
```

```python
df3 = df1.replace('Unknown', np.nan)
df3
```

```python
#distribution of the shooter sex 
df3.Shooter_sex .value_counts(ascending=False)
```

```python
dfData = df3['Shooter_sex'].value_counts().plot(kind='bar',
                                    figsize=(12,6),
                                    title="Distribution of shooter sex ")
dfData.set_xlabel("Shooter_sex")
dfData.set_ylabel("Frequency")
plt.show()
```

From this we can clearly had a first idea, that on the most popular shooter profile based on this dataset it would be a male.

```python
#CHECKING if shooter have mental illness in general.
sns.countplot(data = df3, x = 'General MI')
```

In contrary to what people are thinking we can see that people with confirmed mental illness is almost equal to the number of shooter where we are sure that they did not have any mental illness.

```python
#Shooter race distribution 
sfData = df3['Shooter Race'].value_counts().plot(kind='bar',
                                    figsize=(12,6),
                                    title="Distribution of shooter race")
sfData.set_xlabel("Shooter Race")
sfData.set_ylabel("Frequency")
plt.show()
```

From this graph we can see that most shooter are White american or European American.

```python
# valid float takes in any value, string etc, and returns True if it can convert it to a float
def valid_float(val):
    try:
        float(val)
        return True
    except:
        return False
    
# average of a list of numbers (sum divided by length)
def mean(numbers):
    return sum(numbers)/len(numbers)

# from df3, select only rows where df3['Shooter_age'] is successfully returning True from valid_float
valid_rows = df3[df3['Shooter_age'].apply(lambda age: valid_float(age))]
# replace valid rows with only the selected roles
valid_rows = valid_rows[['Shooter_sex', 'Shooter_age', 'Shooter Race', 'School Related','State']]
valid_rows = valid_rows.dropna()
valid_rows['Shooter_age'] = valid_rows['Shooter_age'].astype(float)
mean(valid_rows['Shooter_age'])
```

thus the average shooter age  is 31.66, which is a very young age.Meaning that youth are definitely involved in this violence
